```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

fig, ax = plt.subplots()
ax.grid()    
fig.tight_layout()

x =  np.arange(1, 30, 1)

y_5 = [poisson.pmf(i, 5)for i in x]
y_10 = [poisson.pmf(i, 10)for i in x]
y_15 = [poisson.pmf(i, 15)for i in x]

plt.plot(x, y_5)
plt.plot(x, y_10)
plt.plot(x, y_15)

```

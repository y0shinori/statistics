```python
import numpy as np
import matplotlib.pyplot as plt

#1時間にlambda匹釣れる釣り堀で、x分以内に魚が釣れる確率
def cum_exp_dist(lambda_, x):
    return 1-np.exp(-lambda_ * x)

x_10 = np.arange(0, 1, 0.01)
y_10 = [cum_exp_dist(10, i)for i in x_10]
x_10 *= 60

x_5 = np.arange(0, 1, 0.01)
y_5 = [cum_exp_dist(5, i)for i in x_5]
x_5 *= 60

x_1 = np.arange(0, 1, 0.01)
y_1 = [cum_exp_dist(1, i)for i in x_1]
x_1 *= 60

fig, ax = plt.subplots()
ax.grid()    
fig.tight_layout()

plt.plot(x_10, y_10, color = 'r',alpha=0.5, label="λ= %d" % 10)
plt.plot(x_5, y_5, color = 'y',alpha=0.5, label="λ= %d" % 5)
plt.plot(x_1, y_1, color = 'b',alpha=0.5, label="λ= %d" % 1)
plt.savefig('hoge.png')

plt.legend()
plt.show()
```
